```python
movie = movies.merge(credits,on='title')
```

```python
movie=movie[['id', 'title', 'overview', 'genres', 'keywords','cast', 'crew']]
```

```python
movies.head()
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace=True)
```

```python
movie.duplicated().sum()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movie['genres']=movie['genres'].apply(convert)
```

```python
movie.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1


    return L
```

```python
movie['cast'] = movie['cast'].apply(convert)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movie['crew']=movie['crew'].apply(fetch_director)
```

```python
movie.head()
```

```python
movie['over']=movie['overview'].apply(lambda x:x.split())
```

```python
movie.head()
```

```python
def collapser(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movie['cast'] = movie['cast'].apply(collapser)
movie['crew'] = movie['crew'].apply(collapser)
movie['genres'] = movie['genres'].apply(collapser)
movie['keywords'] = movie['keywords'].apply(collapser)
```

```python
movie['tags'] = movie['overview'].astype(str) + movie['genres'].apply(lambda x: ' '.join(x)) + movie['keywords'].apply(lambda x: ' '.join(x)) + movie['cast'].apply(lambda x: ' '.join(x)) + movie['crew'].apply(lambda x: ' '.join(x))
movie.head(1)
```

```python
new_df=movie[['id', 'title', 'tags']]
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new_df['tags']).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
cosine_similarity(vector).shape
similarity
```

```python
def recommend(movi):
  movies_index = new_df[new_df['title'] == movi].index[0]
  distances=similarity[movies_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Rockaway')
```
